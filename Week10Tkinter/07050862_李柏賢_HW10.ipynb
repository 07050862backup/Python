### Homework #1 請設計一個 BMI 計算程式。

```python
from tkinter import *

window = Tk()
window.title('BMI App')
window.geometry('800x600')
window.configure(background='white')




def calculate_bmi_number():
    height = float(height_entry.get())
    weight = float(weight_entry.get())
    bmi_value = weight / (height*height)
    result = '你的 BMI 指數是：{} {}'.format(bmi_value, get_bmi_status_description(bmi_value))
    result_label.configure(text=result)

def get_bmi_status_description(bmi_value):
    if bmi_value < 18.5:
        return '，太瘦囉！'
    elif bmi_value >= 18.5 and bmi_value < 24:
        return '體重剛剛好，繼續保持！'
    elif bmi_value >= 24 :
        return '體重有點過重囉，少吃多運動！'




header_label = Label(window, text='BMI 計算器')
header_label.pack()

# 以下為 height_frame 群組
height_frame = Frame(window)
# 向上對齊父元件
height_frame.pack(side=TOP)
height_label = Label(height_frame, text='身高（m）')
height_label.pack(side=LEFT)
height_entry = Entry(height_frame)
height_entry.pack(side=LEFT)

# 以下為 weight_frame 群組
weight_frame = Frame(window)
weight_frame.pack(side=TOP)
weight_label = Label(weight_frame, text='體重（kg）')
weight_label.pack(side=LEFT)
weight_entry = Entry(weight_frame)
weight_entry.pack(side=LEFT)

result_label = Label(window)
result_label.pack()

calculate_btn = Button(window, text='馬上計算',command=calculate_bmi_number)
calculate_btn.pack()

window.mainloop()
```

### Homework #3 請設計一個貸款計算程式，請使用 tkinter 設計此程式。

這個程式的每月支付金額與總支付金額使用淺黃色為背景，我們可以輸入利率、貸款年數、貸款金額然後計算每月支付金額與總支付金額，更多細節可以參考下列執行結果。

```python
from tkinter import * # Import tkinter
    
class LoanCalculator:
    def __init__(self):
        window = Tk() # Create a window
        window.title("Loan Calculator") # Set title
        
        lab1 = Label(window, text = "利率").grid(row = 1,column = 1, sticky = W)
        lab2 = Label(window, text = "貸款年數").grid(row = 2,column = 1, sticky = W)
        lab3 = Label(window, text = "貸款金額").grid(row = 3,column = 1, sticky = W)
        lab4 = Label(window, text = "每月支付金額").grid(row = 4,column = 1, sticky = E)
        lab5 = Label(window, text = "總支付金額").grid(row = 5,column = 1, sticky = E)
        
        self.annualInterestRateVar = StringVar()
        
        e1 = Entry(window, textvariable = self.annualInterestRateVar,justify = LEFT)
        e1.grid(row = 1, column = 2)
        e1.insert(1,"0.0")                # 預設文字方塊1內容
        self.numberOfYearsVar = StringVar()
        
        e2 = Entry(window, textvariable = self.numberOfYearsVar,justify = LEFT)
        e2.grid(row = 2, column = 2)
        e2.insert(1,"0") 
        self.loanAmountVar = StringVar()
        
        e3 = Entry(window, textvariable = self.loanAmountVar,justify = LEFT)
        e3.grid(row = 3, column = 2)
        e3.insert(1,"0") 
        self.monthlyPaymentVar = StringVar()
        
        lblMonthlyPayment = Label(window, textvariable = self.monthlyPaymentVar,bg="lightyellow").grid(row = 4, column = 2)
        self.totalPaymentVar = StringVar()
        lblTotalPayment = Label(window, textvariable = self.totalPaymentVar,bg="lightyellow")
        lblTotalPayment.grid(row = 5,column = 2 )
        btComputePayment = Button(window, text = "計算",command = self.computePayment).grid(row = 6, column = 2, sticky = E)
        
        window.mainloop() # Create an event loop

    def computePayment(self):
        monthlyPayment = self.getMonthlyPayment(
            float(self.loanAmountVar.get()), 
            float(self.annualInterestRateVar.get()) / 1200, 
            int(self.numberOfYearsVar.get()))
        self.monthlyPaymentVar.set(format(monthlyPayment, '10.2f'))
        totalPayment = float(self.monthlyPaymentVar.get()) * 12 \
            * int(self.numberOfYearsVar.get())
        self.totalPaymentVar.set(format(totalPayment, '10.2f'))
        
    def getMonthlyPayment(self,
            loanAmount, monthlyInterestRate, numberOfYears):
        monthlyPayment = loanAmount * monthlyInterestRate / (1
           - 1 / (1 + monthlyInterestRate) ** (numberOfYears * 12))
        return monthlyPayment;

LoanCalculator()  # Create GUI 
```
