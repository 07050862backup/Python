### Homework #1 請使用海龜繪圖繪製六角星星。

```python
import turtle
turtle.clearscreen()
t = turtle.Pen()

for i in range (1,4):
    t.forward(100)
    t.left(120)
    
t.forward(33)
t.left(300)
t.forward(33)

t.left(120)
for i in range (1,4):
    t.forward(100)
    t.left(120)



turtle.exitonclick()
```


### Homework #2 請使用海龜繪圖設計紅綠燈動畫，三顆燈輪流亮起，各亮10秒，循環三次。


```python
import turtle
import time
turtle.clearscreen()

colorsList = ['green', 'yellow', 'red','white']
t1 = turtle.Pen()
t2 = turtle.Pen()
t3 = turtle.Pen()
t4 = turtle.Pen()

changepen = [t1, t2, t3]

t2.penup()
t2.forward(100)
t2.pendown()
t3.penup()
t3.forward(200)
t3.pendown()
t4.penup()
t4.forward(250)
t4.pendown()


for i in range(3):          #畫三個圓
    changepen[i].circle(50)

t4.left(90)
t4.forward(100)
t4.left(90)
t4.forward(300)
t4.left(90)
t4.forward(100)
t4.left(90)
t4.forward(300)
t4.right(90)
t4.forward(300)
for count in range(0,3):
    for i in range(0,3):
        changepen[i].fillcolor(colorsList[i%3])    # 更改色彩
        changepen[i].begin_fill()                  # 開始填充
        changepen[i].circle(50)
        changepen[i].end_fill()
        time.sleep(10)
        changepen[i].fillcolor(colorsList[3])    # 更改色彩
        changepen[i].begin_fill()                  # 開始填充
        changepen[i].circle(50)
        changepen[i].end_fill()                    # 結束填充
```

### Homework #4 請設計一個 Selenium 爬蟲程式，請開啟天瓏書局網站，搜尋字串 "Pythpn"，然後列出搜尋結果。

```python
from selenium import webdriver
import time

browser = webdriver.Chrome()
url = 'https://www.tenlong.com.tw/'
browser.get(url)                # 網頁下載至瀏覽器


txtBox = browser.find_element_by_id('keyword')
txtBox.send_keys('Python')          # 輸入表單資料
# time.sleep(5)                       # 暫停5秒
txtBox.submit()                     # 送出表單



tag5 = browser.find_elements_by_tag_name('img')         # 傳回<img>
for i in range(len(tag5)):
    print("%s " % tag5[i].get_attribute('alt'))
```
