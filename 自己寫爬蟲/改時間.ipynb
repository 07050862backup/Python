```python
res = requests.get("http://eportfolio.mcu.edu.tw/ePortfolio/Student/page01/index.asp?ch=2", headers=headers)

res.encoding = res.apparent_encoding 
html = res.text 
#print(html)

soup = bs4.BeautifulSoup(html, 'lxml')         #lxml是解析HTML文件方式
print(soup)
```

```python
import pandas as pd
 
 
bus_table = pd.read_html(str(soup))
print(bus_table[1])
import numpy as np
boollist = np.zeros((15,7), dtype=int)

def isNaN(num):
    return num != num


for i in range(1,8):
     for j in range(1,16):
        if isNaN(bus_table[1][i][j]) == True:
             boollist[j-1][i-1] = 1
print(boollist)
```
