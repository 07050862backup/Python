```python
import requests
import bs4
import time
res = requests.get("https://github.com/07050862/Adafruit_Python_DHT/network/members")

res.encoding = res.apparent_encoding 

html = res.text 
# 找出repo裡有MCS的字，通常會是我們系的
pattern = 'MCS'
objSoup = bs4.BeautifulSoup(html, 'lxml')         #lxml是解析HTML文件方式
i=1
a_tags = objSoup.select('div.repo a')
for a in a_tags:
    if i % 3 == 2: 
#         print('title', a.text)
        url = 'https://github.com/'+a.text+'?tab=repositories'
        htmlfile = requests.get(url)
        if htmlfile.status_code == requests.codes.ok:
            if pattern in htmlfile.text:                # 方法1
                print(url)
        else:
            url = 'https://github.com/'+a.text+'?tab=repositories'
            while htmlfile.status_code != requests.codes.ok:
                htmlfile = requests.get(url)
                print("no ok") #爬蟲速度過快會出現錯誤，重新爬取直到爬取成功為止!
                print(url)
                if pattern in htmlfile.text:                # 方法1
                    print(url)
#         time.sleep(1)

    i=i+1
```
