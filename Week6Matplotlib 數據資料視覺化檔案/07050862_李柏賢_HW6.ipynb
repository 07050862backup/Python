## Homework #1

在這個專題中以圖書分類的銷售資料在一個圖表區中分別繪製男性、女性銷售率圓餅圖、長條圖及折線圖。

以下是某圖書公司,各分類的男女性銷售資料:

接著想要利用這些資料在同一個圖表區分別分析:

1.書分類男性銷售比率-圓餅

2.書分類女性銷售比率-圓餅圖

3.書分類男女性銷售-長條圖

4.圖書分類男女性銷售-折線圖

```python
import matplotlib.pyplot as plt

sizes_man   = [14, 16, 8, 13, 16, 12, 16, 5]
sizes_woman = [10, 19, 6, 10, 13, 13, 20, 9]
labels  = ["商業理財", "文學小說", "藝術設計", "人文科普", "語言電腦", "心靈養生", "生活風格", "親子共享"]
colors  = ["steelblue", "darkorange", "limegreen", "crimson", "mediumpurple", "sienna", "violet", "gray"]
explode = (0, 0, 0, 0, 0, 0, 0, 0)            ##凸出距離數值串列



plt.figure(figsize=[16,10])

plt.subplot(2,2,1)
plt.pie(sizes_man, 
	explode = explode, 
	labels = labels, 
	textprops = {"fontsize" : 13},
	colors = colors,
	labeldistance = 1.1, 
	autopct = "%2.1f%%", 
	pctdistance = 0.6,
	shadow = False,
	startangle = 360)
# 設定中文字型及負號正確顯示
plt.rcParams["font.sans-serif"] = "Microsoft JhengHei" 
plt.rcParams["axes.unicode_minus"] = False
plt.title("圖書分類銷售比率-男性", fontsize=18)




plt.subplot(2,2,2)
plt.pie(sizes_woman,                
	explode = explode,              #每一個項目凸出距離數字組成的串列,「0」表示正常顯示。
	labels = labels,                #每一個項目標題組成的串列
	textprops = {"fontsize" : 13},  # 文字大小
	colors = colors,                #每一個項目顏色字元組成的字串或是串列,如'rgb'或 ['r','g','b']。
	labeldistance = 1.1,            #項目標題與圓心的距離是半徑的多少倍,例如「1.1」表示項目標題與圓心的距離是半徑的1.1倍。
	autopct = "%2.1f%%",            #項目百分比的格式,語法為「%格式%%」,例如「%2.1f%%」表示整數2位數，小數1位數。
	pctdistance = 0.6,              #百分比文字與圓心的距離是半徑的多少倍。
	shadow = False,                 #布林值，True表示圖形有陰影,False表示圖形沒有陰影。
	startangle = 360)               #開始繪圖的起始角度,繪圖會以逆時針旋轉計算角度。
# 設定中文字型及負號正確顯示
plt.rcParams["font.sans-serif"] = "Microsoft JhengHei" 
plt.rcParams["axes.unicode_minus"] = False
plt.title("圖書分類銷售比率-女性", fontsize=18)



plt.subplot(2,2,3)

width = 0.35                                              ###設定每個項目的寬度(再由子項目分)
listx1 = [x - width/2 for x in range(len(labels))]         ###EX1
listx2 = [x + width/2 for x in range(len(labels))]         ###EX2
plt.bar(listx1, sizes_man, width, label='男')
plt.bar(listx2, sizes_woman, width, label='女')
plt.xticks(range(len(labels)), labels=labels)     ####EX3   
plt.legend(loc = 'upper left')
plt.title("圖書分類銷售長條圖-性別", fontsize=15)
plt.xticks(rotation=50, fontsize=15)         # 傾斜50度
plt.xlabel("圖書分類", fontsize=15)
plt.ylabel("銷售比率(%)", fontsize=15)
# 設定中文字型及負號正確顯示
plt.rcParams["font.sans-serif"] = "Microsoft JhengHei" 
plt.rcParams["axes.unicode_minus"] = False





plt.subplot(2,2,4)

plt.plot( sizes_man, '-s', lw=2, ms=10,label="男")
plt.xticks(range(len(labels)), labels=labels) 
plt.plot( sizes_woman, '-s', lw=2, ms=10,label="女")

plt.legend(loc = 'lower left')
plt.title("圖書分類銷售折線圖-性別", fontsize=15)
plt.xticks(rotation=50, fontsize=15)         # 傾斜50度
plt.xlabel("圖書分類", fontsize=15)
plt.ylabel("銷售比率(%)", fontsize=15)
plt.grid(color='k', ls=':', lw=1, alpha=0.5)



plt.show()
```
