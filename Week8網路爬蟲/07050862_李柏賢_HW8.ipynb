## 2-3.  搜尋網頁特定內容

繼續先前的内容，網頁内容下載後，如果我們想要搜尋特定字串，可以使用許多方法，下列將簡單的用2個方法處理。

```python
import requests
import re

url = 'http://www.mcu.edu.tw'
htmlfile = requests.get(url)
if htmlfile.status_code == requests.codes.ok:
    pattern = input("請輸入欲搜尋的字串 : ")    # pattern存放欲搜尋的字串
    
# 使用方法1
    if pattern in htmlfile.text:                # 方法1
        print("搜尋 %s 成功" % pattern)
    else:
        print("搜尋 %s 失敗" % pattern)
        
# 使用方法2, 如果找到放在串列name內
    name = re.findall(pattern, htmlfile.text)   # 方法2
    if name != None:
        print("%s 出現 %d 次" % (pattern, len(name)))
    else:
        print("%s 出現 0 次" % pattern)
else:
    print("網頁下載失敗")
```

搜尋字串 "銘傳大學" 使用方法1，使用方法2 不僅搜尋，如果找到同時列出執行結果。

這個程式執行時，如果網頁内容下載成功，會要求輸入欲搜尋的字串，將此字串放入 pattemn 變數。

使用2種方法搜尋，方法1 會列出搜尋成功或失敗，方法2 會列出搜尋到此字串的次數。

## 2-4.  下載網頁失敗的異常處理

### 2-4-1 try - except - else

Python在try-except中又增加了else指令，這個指令存放的主要目的是try内的指令正確時，可以執行else内的指令區塊，我們可以將這部分指令區塊稱正確
處理程序,這樣可以增加程式的可讀性。此時語法格式如下:

```python
try:
        指令          #預先設想可能引發異常的指令

except 異常物件:       #若以範例而言,異常物件就是指ZeroDivisionError
        異常處理程序   #通常是指出異常原因,方便修正
        
else:
        正確處理程序   #如果指令正確實執行此區塊指令
```

```python
def division(x, y):
    try:                        # try - except指令
        ans =  x / y
    except ZeroDivisionError:   # 除數為0時執行
        print("除數不可為0")
    else:
        return ans              # 傳回正確的執行結果

print(division(10, 2),"\n")          # 列出10/2
print(division(5, 0),"\n")           # 列出5/0
print(division(6, 3),"\n")           # 列出6/3
```

### 2-4-2 網頁失敗異常處理

有時候我們輸入網址錯誤或是有些網頁有反爬蟲機制,造成下載網頁失敗，Response 物件有 aise_for_status()，可以針對網址正確但是後續檔案名稱錯誤的狀況產生異常處理。

```python
import requests

url = 'http://www.mcu.edu.tw/file_not_existed'  # 不存在的內容
htmlfile = requests.get(url)
try:
    htmlfile.raise_for_status()                 # 異常處理
    print("下載成功")
except Exception as err:                        # err是系統自訂的錯誤訊息
    print("網頁下載失敗: %s" % err)
```

##### ** 很明顯執行異常處理期間又產生了異常，所以程式錯誤產生中斷，有時候可以使用將 requests.get() 放在try:後面。 **

```python
import requests

url = 'http://www.gzaxxc.com/file_not_existed'  # 錯誤的網址
try:
    htmlfile = requests.get(url)
    print("下載成功")
except Exception as err:                        # err是系統自訂的錯誤訊息
    print("網頁下載失敗: %s" % err)
```

上述程式第6行的 raise_for_status() 主要是如果 Response 物件 htmlfile 在前一行擷取網頁內容有錯誤碼時，將可以列出錯誤原因，406錯誤就是網頁伺服器阻擋。用這行程式碼，可以快速中斷協助我們偵錯程式的錯誤。

## 2-6.  爬蟲程式偽成裝瀏覽器

其實我們使用 requests.get() 方法到網路上讀取網頁資料，這類的程式就稱網路爬蟲程式，甚至你也可以將各大公司所設計的搜尋引擎稱為網路爬蟲程式。

為了解決爬蟲程式被伺服器阻擋的困擾，我們可以將所設計的爬蟲程式偽裝成瀏覽器，方法是在程式前端加上 headers 内容。

```python
import requests

headers = { 'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64)\
            AppleWebKit/537.36 (KHTML, like Gecko) Chrome/45.0.2454.101\
            Safari/537.36', }

url = 'http://aaa.24ht.com.tw/'
htmlfile = requests.get(url, headers=headers)
htmlfile.raise_for_status()
print("儰裝瀏覽器擷取網路資料成功")
```

上述的重點是第3-5行的敘述，其實這是一個標題(headers)宣告，第3和4行末端的反斜線 "\" 主要表達下一行與這一行是相同敘述，也就是處理同一敘述太長時分行撰寫，Python會將3-5行視為同一敘述。

然後第8行呼叫 requests.get() 時，第2個參數需要加上 "headers=headrs"，這樣這個程式就可以偽裝成瀏覽器，可以順利取得網頁資料了。

## 2-7.  儲存下載的網頁

使用 requests.get() 獲得網頁内容時，是儲存在 Rsponse 物件類型內，如果要將這類型的物件存入硬碟内，需使用 Response 物件的 iter_content() 方法。

這個方法是採用重複迭代方式將 Respons 物件內容寫入指定的檔案內，每次寫入指定磁區大小是以Bytes為單位，一般可以設定 1024*5 或 1024*10 或更多。

```python
import requests

url = 'http://www.mcu.edu.tw'                     # 網址
try:
    htmlfile = requests.get(url)
    print("下載成功")
except Exception as err:                                # err是系統自訂的錯誤訊息
    print("網頁下載失敗: %s" % err)
# 儲存網頁內容
fn = 'out23_11.txt'
with open(fn, 'wb') as file_Obj:                        # 以二進位儲存
    for diskStorage in htmlfile.iter_content(10240):    # Response物件處理
        size = file_Obj.write(diskStorage)              # Response物件寫入
        print(size)                                     # 列出每次寫入大小
    print("以 %s 儲存網頁HTML檔案成功" % fn)
```

由於這個網頁檔案内容比較大，所以將每次寫入檔案大小設為 10240 bytes，程式第12行所開啟的是以二進位可寫入 "wb" 方式開啟，這是為了怕網頁内有Unicode碼。

程式第13-15行是一個迥圏，這個迴圈會將 Response 物件 htmlfile 以迴圏方式寫入所開啟的 file_Obj，最後是存入第11行設定的 out23_11.txt 檔案内。

程式使用 write() 寫入 Response 物件時會回傳所寫入網頁内容的大小，所以15行會列出當次迴圈所寫入的大小。

## 4-6.  傳回所找尋所有符合的標籤 find_all()

## 4-7.  認識 HTML 元素内容屬性與 getText()

HTML元素內容的屬性有下列3種。

+ textContent：内容，不含任何標籤碼。

+ innerHML：元素內容，含子標籤碼，但是不含本身標籖碼。

+ outerHML：元素內容，含子標籤碼，也含本身標籤碼。

如果有一個元素内容如下:

< p > Marching onto the path of < b > Web Design Expert < /b >< /p >

則上述3個屬性的觀念與内容分別如下:

textContent : Web Design Expert

innerHTML : Marching onto the path of < b > Web Design Expert < /b >

outerHTML : < p > Marching onto the path of < b > Web Design Expert < /b >< /p >

當使用 BeautifulSoup 模組解析 HTML 文件，如果傳回是串列時，也可以配合索引應用 getText() 取得串列元素内容，所取得的內容是 textContent。

意義與 4-4 節的 text 屬性相同。

```python
import bs4

htmlFile = open('myhtml.html', encoding='utf-8')
objSoup = bs4.BeautifulSoup(htmlFile, 'lxml')
objTag = objSoup.find_all('h1')
print("資料型態    = ", type(objTag))     # 列印資料型態
print("列印Tag串列 = ", objTag)           # 列印串列

print("\n使用Text屬性列印串列元素 : ")
for data in objTag:                       # 列印串列元素內容
    print(data.text)
    
print("\n使用getText()方法列印串列元素 : ")
for i in range(len(objTag)):
    print(objTag[i].getText())
```

## 4-8.  select()

select() 主要是以 CSS 選擇器(selector)的觀念尋找元素，如果找到回傳的是串列(list)，如果找不到則傳回空串列。下列是使用實例:

+ objSoup.select('p'):找尋所有< p >標籤的元素。
    
+ objsoup.select('img'):找尋所有 < img >標籤的元素。
    
+ objsoup.select('.happy'):找尋所有 CSS class 屬性為 happy 的元素。

+ objsoup.select('#author'):找尋所有 CSS id 屬性為 author 的元素。

+ objSoup.select('p #author'):找尋所有 < p > 且id 屬性為 author 的元素。

+ objSoup.select('p .happy'):找尋所有 < p > 且 class 屬性為 happy 的元素。

+ objSoup.select('div strong'):找尋所有在 < section > 元素内的 < strong > 元素。

+ objsoup.select('div > strong'):所有在 < section > 内的 < strong > 元素，中間沒有其他元素。

+ objSoup.select('input[name]') :找尋所有 < input > 標籤且有 name 屬性的元素。

HTML文件中常常可以看到標籤内有子標籤，如果查看 myhtml.html 的第21行，可以看到 < p > 標籤内有 < strong > 標籤，碰上這種狀況若是列印串列元素内容時，可以看到子標籤存在。

```python
import bs4

htmlFile = open('myhtml.html', encoding='utf-8')
objSoup = bs4.BeautifulSoup(htmlFile, 'lxml')
pObjTag = objSoup.select('p')
print("含<p>標籤的串列長度 = ", len(pObjTag))
for i in range(len(pObjTag)):
    print(str(pObjTag[i]))              # 內部有子標籤<strong>字串
    print(pObjTag[i].getText())         # 沒有子標籤
    print(pObjTag[i].text)              # 沒有子標籤
```

## 4-9.  標籤字串的 get()

假設我們現在搜尋 < img > 標籤，請參考下列實例。

```python
import bs4

htmlFile = open('myhtml.html', encoding='utf-8')
objSoup = bs4.BeautifulSoup(htmlFile, 'lxml')
imgTag = objSoup.select('img')
print("含<img>標籤的串列長度 = ", len(imgTag))
for i in range(len(imgTag)):              
    print(imgTag[i])    
```

< img > 是一個插入圖片的標籤，沒有結束標籤，所以沒有内文，如果讀者嘗試使用 text 屬性列印内容 "print(imgtag[0].text)" 將看不到任何結果。

< img >對網路爬蟲設計是很重要，因為可以由此獲得網頁的圖檔資訊。從上述執行結果可以看到對我們而言很重要的是 < img > 標籤内的屬性 src ，這個屬性設定了圖片路徑。

這個時候我們可以使用標籤字串的 get() 取得。

```python
import bs4

htmlFile = open('myhtml.html', encoding='utf-8')
objSoup = bs4.BeautifulSoup(htmlFile, 'lxml')
imgTag = objSoup.select('img')
print("含<img>標籤的串列長度 = ", len(imgTag))
for i in range(len(imgTag)):              
    print("列印標籤串列 = ", imgTag[i])
    print("列印圖檔     = ", imgTag[i].get('src'))
```

上述程式最重要是第10行的 imgTag[i].get('src')，這個方法可以取得標籖字串的屬性內容。

曾經說明標籤字串與純字串(str)不同是在這裡，純字串無法呼叫 get() 方法執行上述將圖檔字串取出。

### Homework #1 請使用webbrowser() 開啟T客邦網站，搜尋任意關鍵字，計算出現次數。

```python
import webbrowser
webbrowser.open('https://www.techbang.com')

import requests
import re

url = 'https://www.techbang.com'
htmlfile = requests.get(url)
if htmlfile.status_code == requests.codes.ok:
    pattern = input("請輸入欲搜尋的字串 : ")    # pattern存放欲搜尋的字串
    
# 使用方法1
    if pattern in htmlfile.text:                # 方法1
        print("搜尋 %s 成功" % pattern)
    else:
        print("搜尋 %s 失敗" % pattern)
        
# 使用方法2, 如果找到放在串列name內
    name = re.findall(pattern, htmlfile.text)   # 方法2
    if name != None:
        print("%s 出現 %d 次" % (pattern, len(name)))
    else:
        print("%s 出現 0 次" % pattern)
else:
    print("網頁下載失敗")
```
